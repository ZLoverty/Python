```python
import numpy as np
import trackpy as tp
import pandas as pd
from myImageLib import dirrec, to8bit
from nd2reader import ND2Reader
import os
import pdb
import matplotlib.pyplot as plt
import matplotlib as mpl
import skimage.io
import h5py
from IPython.display import clear_output, display

def bestcolor(n):
    r = [57, 218, 62, 204, 83, 107, 146, 148]
    g = [106, 124, 150, 37, 81, 76, 36, 139]
    b = [177, 48, 81, 41, 84, 154, 40, 61]
    return (r[n]/255, g[n]/255, b[n]/255)

def compile_log(mainFolder, **kwargs):
    save = False
    for kw in kwargs:
        if kw == 'save':
            saveDir = kwargs[kw]
            save = True
    logDir = []
    for folder in mainFolder:
        logDir += dirrec(folder, 'log.csv')
    print(logDir)
    compLog = pd.DataFrame()
    for log in logDir:
        folder, fileName = os.path.split(log)
        folder, date = os.path.split(folder)
        logData = pd.read_csv(log).assign(Date=date)
        logData.rename(mapper={'Unnamed: 0': 'No'}, axis=1, inplace=True)
        logData = logData.loc[:, ['No', 'FPS', 'MPP', 'N', 'CT', 'Chirality', 'Date']]
        logData = logData.where((logData.Chirality=='L') | (logData.Chirality=='R'))
        logData = logData.dropna()
        compLog = compLog.append(logData, ignore_index=True)
    compLog.No = compLog.No.astype('uint8')
    if save == True:
        compLog.to_hdf(saveDir, 'Log')
    return compLog

def compile_traj(mainFolder, log, saveDir): 
    unprocData = pd.DataFrame()
    compiledData = pd.DataFrame()
    for folder in mainFolder:
        for num, row in log.iterrows():    
            with h5py.File(saveDir, 'a') as f:
                col_node = '%08s' % row.Date + '/%02d' % row.No + '/traj/columns'
                val_node = '%08s' % row.Date + '/%02d' % row.No + '/traj/values'
                if (col_node in f) and (val_node in f):
                    continue
                trajDir = os.path.join(folder, '%08s' % row.Date, '%02d' % row.No, 'tracking.csv')
                if os.path.exists(trajDir) == True:
                    traj = pd.read_csv(trajDir)
                    traj.rename(mapper={'Slice': 'frame', 'X': 'x', 'Y': 'y'}, axis=1, inplace=True)
                    traj = traj.loc[:, ['x', 'y', 'Angle', 'Major', 'Minor', 'Area', 'frame', 'particle']]        
                    compiledData = compiledData.append(row.to_frame().transpose())
                else:
                    unprocData = unprocData.append(row.to_frame().transpose())
                    continue
                columns = np.array(traj.columns, dtype=h5py.special_dtype(vlen=str))
                values = traj.values
                f.create_dataset(col_node, data=columns)
                f.create_dataset(val_node, data=values)
                print('{:8s}-{:02}'.format(row.Date, row.No))
    print('Compiled data')
    print(compiledData)
    print('Unprocessed data')
    print(unprocData)

def query_tracking_status(mainFolder, log):
    stat_log = log.assign(finding='X', tracking='X')
    for folder in mainFolder:
        for num, row in stat_log.iterrows():
            dataDir = os.path.join(folder, '%08s' % row.Date, '%02d' % row.No)
            if os.path.exists(os.path.join(dataDir, 'finding.csv')):
                stat_log.at[num, 'finding'] = 'V'
            if os.path.exists(os.path.join(dataDir, 'tracking.csv')):
                stat_log.at[num, 'tracking'] = 'V'
    return stat_log

def construct_multiparticle_trajectory(dataList):
    """
    dataList is a list of DataFrame containing traj's
    """
    mul_traj = pd.DataFrame()
    pCount = 1    
    for data in dataList:
        data1 = data.copy()
        for num in data.particle.drop_duplicates():
            data.particle.loc[data1.particle==num] = pCount            
            pCount += 1
        mul_traj = mul_traj.append(data)
    return mul_traj

class traj_data(h5py._hl.files.File):
    def __init__(self, h5Dir, mode):
        super().__init__(h5Dir, mode)
    def get_traj(self, date, No):
        # date is str 'mmddyyyy'
        # No is str '%02d'
        col_node = date + '/' + No + '/traj/columns'
        val_node = date + '/' + No + '/traj/values'
        if (col_node not in self) or (val_node not in self):
            ValueError('Node ' + col_node + ' does not exist.')            
        col = self[col_node][:]
        val = self[val_node][:]
        traj = pd.DataFrame(columns=col, data=val)
        return traj
    def set_log(self, log):
        # log is a DataFrame table containing parameter info of trajs
        self.log = log
    def get_log(self):
        return self.log
    def cal_msd(self):
        for num, row in self.log.iterrows():
            print('Processing %s - %d' % (row.Date, row.No))
            col_node = row.Date + '/' + '/%02d' % row.No + '/MSD/columns'
            val_node = row.Date + '/' + '/%02d' % row.No + '/MSD/values'
            if (col_node not in self) or (val_node not in self):
                try:
                    data = self.get_traj(row.Date, '%02d' % row.No)
                except:
                    print('Fail loading trajectory %s - %d' % (row.Date, row.No))
                    continue
                data.rename(mapper={'Slice': 'frame', 'X': 'x', 'Y': 'y'}, axis=1, inplace=True)
                try:
                    msd = tp.emsd(data, row.MPP, row.FPS, max_lagtime=int(data.frame.max()))
                except:
                    print('Fail calculating MSD %s - %d' % (row.Date, row.No))
                    continue
                lagt = np.array(msd.index)
                msdv = msd.values
                msdData = pd.DataFrame().assign(lagt=lagt, msd=msdv)
                columns = np.array(msdData.columns, dtype=h5py.special_dtype(vlen=str))
                values = msdData.values
                self.create_dataset(col_node, data=columns)
                self.create_dataset(val_node, data=values)
    def show_attr(self):
        # print all attr's of h5 file
        pass
    def get_msd(self, date, No):
        col_node = date + '/' + No + '/MSD/columns'
        val_node = date + '/' + No + '/MSD/values'
        if (col_node not in self) or (val_node not in self):
            ValueError('Node ' + col_node + ' does not exist.')            
        col = self[col_node][:]
        val = self[val_node][:]
        msd = pd.DataFrame(columns=col, data=val)
        return msd
    def av_msd(self, **kwargs):
        dataList = []
        sublog = self.log.copy()
        for kw in kwargs:
            sublog = sublog.loc[sublog[kw]==kwargs[kw]]
        for num, row in sublog.iterrows():
            try:
                traj = self.get_traj(row.Date, '%02d' % row.No)
            except:
                continue
            dataList.append(traj)
        mul_traj = construct_multiparticle_trajectory(dataList)
        mul_traj.rename(mapper={'Slice': 'frame', 'X': 'x', 'Y': 'y'}, axis=1, inplace=True)
        msd = tp.emsd(mul_traj, mpp=0.33, fps=2, max_lagtime=3000)
        lagt = np.array(msd.index)
        msdv = msd.values
        msdData = pd.DataFrame().assign(lagt=lagt, msd=msdv)
        print(len(mul_traj.particle.drop_duplicates()))
        return msdData
    def check_dup_traj(self):
        dupTrajRec = pd.DataFrame()
        for num, row in self.log.iterrows():
            try:
                data = self.get_traj(row.Date, '{:02}'.format(row.No))
                print('Processing {0}-{1:02}'.format(row.Date, row.No))
            except:
                continue
            for frame in data.Slice.drop_duplicates():
                frameData = data.loc[data.Slice==frame]
                for particle in frameData.particle.drop_duplicates():
                    particleData = frameData.loc[frameData.particle==particle]
                    if len(particleData) > 1:
                        rec = pd.DataFrame(data=[[row.Date, row.No, frame, particle, len(particleData)]],
                                          columns=['date', 'No', 'frame', 'particle', 'numTraj'])
                        dupTrajRec = dupTrajRec.append(rec)
        if len(dupTrajRec) == 0:
            print('No duplicated trajectories')
        else:
            print('{0} errors spotted'.format(len(dupTrajRec)))
            return dupTrajRec
    def get_traj_mul(self, **kwargs):
        dataList = []
        sublog = self.log.copy()
        for kw in kwargs:
            sublog = sublog.loc[sublog[kw]==kwargs[kw]]
        for num, row in sublog.iterrows():
            try:
                traj = self.get_traj(row.Date, '%02d' % row.No)
            except:
                continue
            dataList.append(traj)
        mul_traj = construct_multiparticle_trajectory(dataList)
        return mul_traj
```

```python
def cal_D(msd, plot=False):
    fit_range = 0.6  
    sample_number = 10
    logmsd = np.log(msd)
    lagt = logmsd['lagt']
    total_range = lagt.max() - lagt.min()
    starts = np.linspace(lagt.min(), lagt.max()*(1-fit_range), sample_number)
    fit_result = []
    for start in starts:
        sample = logmsd.loc[(logmsd.lagt>=start)&(logmsd.lagt<start+total_range*fit_range)]
        p = np.polyfit(sample['lagt'], sample['msd'], 1, full=True)
        fit_result.append(p[0])
        if plot == True:
            x = np.linspace(sample['lagt'].min(), sample['lagt'].max(), 20)
            y = p[0][0]*x + p[0][1]
            x = np.exp(x)
            y = np.exp(y)
            plt.plot(x, y)
    fit_result = np.array(fit_result)
    fit_result = pd.DataFrame(data=fit_result, columns=['slope', 'intersect'])
    intersect = fit_result.intersect.iloc[abs(fit_result.slope-1).idxmin()]
    D = 0.25*np.exp(intersect)
    return D 
```

```python
plt.figure(figsize=(10,7))
c = [5, 10, 20]
L = [cal_D(L5), cal_D(L10), cal_D(L20)]
R = [cal_D(R5), cal_D(R10), cal_D(R20)]
plt.plot(c, L, label='Left-handed', ls='--', marker='o')
plt.plot(c, R, label='Right-handed', ls='--', marker='o')
plt.ylabel(r'$D_T$ (um$^2$/s)')
plt.xlabel(r'Concentration (n$_0$)')
plt.savefig('D_conc.png')
```

```python
def plot_pdf(traj, bins=20, ti=[1, 2, 3, 4]):
    plt.figure(figsize=(10, 10))
#     ti = list(range(4, 40, 8))
    for dt in ti:
        numbers = traj.particle.drop_duplicates()
        dx_all = pd.DataFrame()
        for no in numbers:
            traj1 = traj.loc[traj.particle==no]
            dx = traj1.y.diff(dt)
            dx = dx.dropna()
            if len(dx_all) == 0:
                dx_all = dx
            else:
                dx_all = dx_all.append(dx)
        hist, bin_edges = np.histogram(dx_all, bins='scott', density=True)
        binsize = bin_edges[1] - bin_edges[0]
        plt.plot(bin_edges[:-1]+binsize/2, hist, marker='o', ls='-', markersize=5, mfc='none', label='$\Delta t$ = {0} s'.format(dt/10))
    plt.legend(frameon=False)
    plt.xlabel('$\Delta x$', fontsize=20)
    plt.ylabel('Displacement PDF', fontsize=20)
#     plt.title('%s particle, %d n$_0$' % (Chirality, N))    
```

```python
def plot_hist(traj, fps=10, bins=20, ti=[1, 2, 3, 4]):
    plt.figure(figsize=(10, 10))
#     ti = list(range(4, 40, 8))
    for dt in ti:
        numbers = traj.particle.drop_duplicates()
        dx_all = pd.DataFrame()
        for no in numbers:
            traj1 = traj.loc[traj.particle==no]
            dx = traj1.y.diff(dt)
            dx = dx.dropna()
            if len(dx_all) == 0:
                dx_all = dx
            else:
                dx_all = dx_all.append(dx)
        hist, bin_edges = np.histogram(dx_all, bins=bins)
        binsize = bin_edges[1]-bin_edges[0]
        plt.plot(bin_edges[:-1]+binsize/2, hist, marker='o', ls='-', markersize=5, mfc='none', label='$\Delta t$ = {0} s'.format(dt/fps))
    plt.legend(frameon=False)
    plt.xlabel('$\Delta x$', fontsize=20)
    plt.ylabel('Hist', fontsize=20)
#     plt.title('%s particle, %d n$_0$' % (Chirality, N))    
```

```python
def draw_msd_eb(ax, log, color, label, **kwargs):    
    with traj_data(r'I:\Google Drive\Research projects\C\Data\CompData\New\traj.h5', 'r') as f:
        f.set_log(log)
        sublog = log
        for kw in kwargs:
            sublog = sublog.loc[sublog[kw]==kwargs[kw]]
        print(sublog)
        count = 0
        for num, row in sublog.iterrows():
            msd = f.get_msd(row.Date, '%02d'% row.No)
            msd.index = msd['lagt']
            msd = msd.drop(columns='lagt')
#             msd = msd.rename(mapper={'msd':'msd'+str(count)})
            if count == 0:
                msdt = msd
            else:
                msdt = msdt.join(msd, lsuffix='', rsuffix=str(count))
            count += 1
#         msdt = msdt.dropna()
        av = msdt.mean(axis=1, skipna=True)
        std = msdt.std(axis=1, skipna=True)
        ax.errorbar(msdt.index.values, av, yerr=std, ls='', marker='s', color=color, label=label)
    return msdt.assign(av=av, std=std)
```
