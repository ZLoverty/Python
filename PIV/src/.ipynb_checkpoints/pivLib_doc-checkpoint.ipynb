### Perform PIV analysis on an image sequence
<code>imseqPIV(folder, winsize, overlap, dt)</code><br><br>
* folder: directory of the folder where image sequence is stored
* winsize: PIV window size (pix)
* overlap: overlap between adjacent boxes (pix)
* dt: time interval between two frames (s)

Return:
* A pd.DataFrame containing columns ('x', 'y', 'u', 'v', 'frame')

Note:
* A smoothening has been embedded in this function.

```python
from pivLib import imseqPIV
folder = r'I:\Google Drive\Code\Python\PIV\test_images\imseq'
winsize = 50 # pixels
overlap = 25 # pixels
dt = 0.033 # frame interval (sec)
data = imseqPIV(folder, winsize, overlap, dt)
```

```python
data.head()
```

```python
# Visualize PIV result
data1 = data.loc[data.frame==1]
plt.figure(figsize=(10, 10))
plt.quiver(data1.x, data1.y, data1.u, data1.v)
plt.axis('off')
plt.axis('equal')
```
