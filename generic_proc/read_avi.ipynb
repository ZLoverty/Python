# Extract snapshots from .avi videos

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from corrLib import readdata
import os
```

```python
# read video
cap = cv2.VideoCapture('/home/zhengyang/data/DE/07082021/300-300-42_C001H001S0001/300-300-42_C001H001S0001.avi')
count = 0
while True:
    ret, frame = cap.read()
    if ret:
        print('frame {:04d}'.format(count))
        plt.imshow(frame)        
        count += 1
    else:
        break                
    if count == 1:
        break
```

```python
folder = '/home/zhengyang/data/DE/07082021/'
save_folder = '/home/zhengyang/data/DE/07082021/snapshots/'
l = readdata(folder, 'avi')
```

```python
for num, i in l.iterrows():
    cap = cv2.VideoCapture(i.Dir)
    ret, frame = cap.read()
    save_dir = os.path.join(save_folder, i.Name + '.jpg')
    _folder = os.path.split(save_dir)[0]
    if os.path.exists(_folder) == False:
        os.makedirs(_folder)
    cv2.imwrite(save_dir, frame)
    print(save_dir)
```
