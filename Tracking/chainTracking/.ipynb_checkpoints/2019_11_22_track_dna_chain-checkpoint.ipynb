```python
import os
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd
from pandas import DataFrame, Series
import pims
import pims_nd2
import trackpy as tp
from pims import ND2_Reader
from skimage import io, exposure, util, measure, feature, filters, data, morphology, color, transform, draw, segmentation
from scipy import ndimage, misc, signal, spatial, stats, fftpack
import cv2
from matplotlib import animation, rc, quiver
import matplotlib.patches as mpatches
from matplotlib import collections as mc
from IPython.display import HTML
import trackpy as tp
import sys
from mpl_toolkits import mplot3d
#from sklearn.preprocessing import normalize
from sklearn import preprocessing
import seaborn as sns
%matplotlib inline
```

```python
loadf = ND2_Reader('R://Dip/DNA_chain/2kbp_10ug_per_ml_1.nd2')
frames = np.asarray(loadf)
```

```python
image = frames[0][400:800, 400:800]

plt.imshow(image)
```

```python
thresh = filters.threshold_otsu(image)
thresh_image = image > thresh
plt.imshow(thresh_image)
```

```python
image_g = filters.gaussian(image, 1)
seed = np.copy(image_g)
seed[1:-1, 1:-1] = image_g.min()
mask = image_g
dilated = morphology.reconstruction(seed, mask, method='dilation')
fig, ax = plt.subplots()
ax.imshow(dilated, cmap='gray')
```

```python
#plt.imshow(dilated)
dilated_sharp = 1/(dilated**(1/2))
plt.imshow(dilated_sharp, cmap='gray')
#sns.distplot(dilated.flatten())
```

```python
thresh_filt = filters.threshold_otsu(dilated_sharp)
bw_dilsh = dilated_sharp > thresh_filt
plt.imshow(morphology.remove_small_objects(bw_dilsh, min_size=100)*dilated_sharp, cmap = 'gray')
```

```python
local_maxima = morphology.extrema.h_maxima(dilated_sharp, h = 0.05)
label_maxima = measure.label(local_maxima)
#overlay = color.label2rgb(label_maxima, dilated_sharp, alpha=0.7, bg_label=0,
#                          bg_color=None, colors=[(1, 0, 0)])
plt.imshow(label_maxima)
```

```python
thresh_d = filters.threshold_otsu(dilated)
thresh_image_d_uf = dilated > thresh_d
thresh_image_d = morphology.remove_small_holes(thresh_image_d_uf, min_size=400)
plt.imshow(thresh_image_d)
```

```python
dist_d = ndimage.distance_transform_edt(thresh_image_d)
local_maxi = feature.peak_local_max(dist_d, indices=False, footprint=np.ones((3, 3)),
                            labels=thresh_image_d)
fig, ax = plt.subplots()
ax.imshow(-dist_d)
```

```python
local_maxima = morphology.extrema.local_minima(thresh_image_d)
label_maxima = measure.label(local_maxima)
overlay = color.label2rgb(label_maxima, thresh_image_d, alpha=0.7, bg_label=0,
                          bg_color=None, colors=[(1, 0, 0)])
```

```python
im_max = ndimage.maximum_filter(thresh_image_d, size = 20, mode = 'constant')

coordinates = feature.peak_local_max(util.invert(thresh_image_d), min_distance=100)
#plt.imshow(im_max)
plt.plot(coordinates[:, 1], coordinates[:, 0], 'r.')
```

```python
markers = ndimage.label(local_maxi)[0]
plt.imshow(markers)
```
