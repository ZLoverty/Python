```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
case = np.array([1, 2, 5, 6, 7, 11, 12, 13, 14, 15, 16, 17, 35, 53, 57, 59, 60, 66,
       71, 91, 109, 130, 161, 236, 340, 456, 577, 757, 1047, 1330, 1767, 2053])
day = np.array(range(1, len(case)+1))
```

```python
def power(x, a, b):
    return a * x ** b
```

```python
x1 = day[day<=10]
y1 = case[day<=10]
po1, pv1 = curve_fit(power, x1, y1)
x2 = day[day>10]
y2 = case[day>10]
po2, pv2 = curve_fit(power, x2, y2)
```

```python
plt.plot(day, case)
plt.plot(x1, power(x1, *po1)*2)
plt.text(x1.mean(), 20, '{:.2f}'.format(po1[1]))
plt.plot(x2, power(x2, *po2)*2)
plt.text(x2.mean(), 1000, '{:.2f}'.format(po2[1]))
plt.xscale('log')
plt.yscale('log')
```
