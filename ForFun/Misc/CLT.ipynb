```python
import numpy as np
import matplotlib.pyplot as plt
from miscLib import label_slope
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['figure.figsize'] = 10, 6
matplotlib.rcParams['font.family'] = 'serif'
```

```python
n = 100
```

```python
def pn(n):
    d = 0
    for i in range(0, n+1):
        d += factorial(n) / factorial(i)/factorial(n-i)/2**n*abs(i-n/2)
    return d
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact
```

```python
nL = range(10, 100, 10)
N = []
dN = []
for n in nL:
    N.append(n / 2)
    dN.append(pn(n)) 
    
```

```python
x = np.array(N)
y = np.array(dN)
plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$N$')
plt.ylabel('$\Delta N$')
xf, yf, xt, yt, slope = label_slope(x, y, location='n')
plt.plot(xf, yf, ls='--', color='black')
plt.text(xt, yt, '{:.1f}'.format(slope))
plt.xscale('log')
plt.yscale('log')
```
