## xf, yf, xt, yt, slope = label_slope(x, y, location='n')
x, y represent a curve that follows power-law-like relation. label_slope(x, y, location='n') takes in the data of the curve and returns 
* (xf, yf): the position of slope-indicator dashed line
* (xt, yt): position of slope text 
* slope : slope value of dashed line

```python
# Read example data
data = pd.read_csv(r'I:/Github/Python/Correlation/test_images/boxsize_effect/density_fluctuation_data.csv')
x = data.N
y = data.dN
```

```python
xf, yf, xt, yt, slope = label_slope(x, y, location='n')
```

```python
plt.plot(x, y)
plt.plot(xf, yf, ls='--', color='black')
plt.text(xt, yt, '{:.2f}'.format(slope))
plt.xscale('log')
plt.yscale('log')
```
