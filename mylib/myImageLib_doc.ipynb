### dirrec(path, filename)
Recursively look for files specified by *filename* under folder *path*.

Arguments
* path: the directory where you want to look for files.
* filename: name of the files you want to look for.

return
* a list of full directories of files with name filename

Note
* filename can be partially specified, e.g. '*.py' to search for all the         .py files or 'track*' to search for all files starting with 'track'.

Example

```python
path = r'I:\Github\Python\Correlation\test_images\df_result'
mil.dirrec(path, '*.csv')
```

### to8bit(img)
Enhance contrast and convert to 8-bit

Arguments
* img: raw image which is not 8-bit

return
* an 8-bit image with enhanced contrast

Example

```python
imgDir = r'I:\Github\Python\generic_proc\test_images\40-2.tif'
img = io.imread(imgDir)
img8 = mil.to8bit(img)
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap='gray')
ax[0].set_title('img')
ax[1].imshow(img8, cmap='gray')
ax[1].set_title('img8')
```

### bpass(img, low, high)

Arguments
* img: an 8-bit mono image
* low: low pass, the variation with wavelength smaller than *low* will be removed
* high: high pass, the variation with wavelength larger than *high* will be removed

return
* bandpassed image, usually the illumination inhomogeneity is removed

Example

```python
imgDir = r'I:\Github\Python\generic_proc\test_images\40-2.tif'
img = io.imread(imgDir)
img8 = mil.to8bit(img)
bp = mil.bpass(img8, 3, 100)
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img8, cmap='gray')
ax[0].set_title('img')
ax[1].imshow(bp, cmap='gray')
ax[1].set_title('bpass')
```

### bestcolor(n)



Arguments
* n: color number, there are 10 color total in this list, so n<10. The ten colors are chosen according to [Optimal Colors for Graphs](http://ksrowell.com/blog-visualizing-data/2012/02/02/optimal-colors-for-graphs/)

Return
* A color that is good for make multiline plots

Colormap

```python
x = range(0, 10)
y = np.zeros((1, 10)).flatten()
k = 0
plt.figure(figsize=(6, 1))
for xx, yy in zip(x, y):
    plt.plot(xx, yy, color=mil.bestcolor(k), marker='s', markersize=40)
    plt.text(xx-0.2, yy-0.003, str(k), fontdict={'size': 20})
    k += 1
plt.axis('off')
```

### wowcolor(n)

Arguments
* n: color number, there are 22 color total in this list, so n<22. The 22 colors are chosen according to [gamepedia](https://wow.gamepedia.com/Class_colors) and [Optimal Colors for Graphs](http://ksrowell.com/blog-visualizing-data/2012/02/02/optimal-colors-for-graphs/). The first 12 colors are class colors in WoW, the latter 10 colors are the same as those in bestcolors(n).

Return
* A color 

Colormap

```python
x = range(0, 22)
y = np.zeros((1, 22)).flatten()
k = 0
plt.figure(figsize=(12, 1))
for xx, yy in zip(x, y):
    plt.plot(xx, yy, color=mil.wowcolor(k), marker='s', markersize=40)
    plt.text(xx-0.4, yy-0.003, str(k), fontdict={'size': 14})
    k += 1
plt.axis('off')
```
