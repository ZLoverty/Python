# Brownian particles

100 particles uniformly distributed in 2-D space.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
x = np.arange(0, 20)
X, Y = np.meshgrid(x, x)
```

```python
plt.figure(dpi=300, figsize=(3, 3))
plt.scatter(X, Y, s=3)
```

generate 20 frames with Gaussian random noise on each particle.

```python
folder = r'E:\Google Drive\Research projects\DF\pictures&videos\particle-eq-noneq\non-eq'
for i in range(0, 20):
    xnoise = 20*np.random.rand(*X.shape)
    ynoise = 20*np.random.rand(*X.shape)
    plt.figure(dpi=300, figsize=(3, 3))
    plt.scatter(X+xnoise, Y+ynoise, s=3)
    plt.axis('off')
    plt.savefig(os.path.join(folder, '{:02d}.jpg'.format(i)))
```
