# Trial of using sonication to destroy structure in bacteria (-)

* Compare density fluctuation curves between pre- and post- sonication for each sample, 10 in total. 

```python
import os
from skimage import io
import corrLib as cl
import matplotlib.pyplot as plt
from myImageLib import wowcolor
import matplotlib
import numpy as np
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['figure.figsize'] = 10, 6
matplotlib.rcParams['font.family'] = 'serif'
```

```python
folder = r'D:\Data\02182020\img'
sfL = next(os.walk(folder))[1]
for sf in sfL:
    img = io.imread(os.path.join(folder, sf, '16-bit', '0000.tif'))
```

```python
img0 = io.imread(os.path.join(folder, sfL[0], '16-bit', '0000.tif'))
img1 = io.imread(os.path.join(folder, sfL[3], '16-bit', '0000.tif'))
d0 = cl.density_fluctuation(img0)
d1 = cl.density_fluctuation(img1)
plt.plot(d0.n, d0.d/d0.n**0.5, label='raw', ls='', marker='s',
            mfc=(0,0,0,0), mec=wowcolor(0), markersize=10)
plt.plot(d1.n, d1.d/d1.n**0.5, label='sonicate', ls='', marker='s',
            mfc=(0,0,0,0), mec=wowcolor(1), markersize=10)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.xlabel('$l^2/l_b^2$')
plt.ylabel('$\Delta N/\sqrt{N}$')
```

__Comment__: Sonication (for time ranging from 1s to 10s) does not show obvious effect on dense bacterial suspension micro-structure. In 5 cases out of 10 total samples, the number fluctuations get more pronounced after sonication, indicating that sonication may not be able to make the suspensions homogeneous. Instead, it causes more aggregation and thus more inhomogeneity. This observation is consistent with previous experiment, where I dipped the whole slide into sonicator for 10s.
