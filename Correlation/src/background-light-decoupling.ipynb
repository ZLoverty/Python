The microscope illumination light has intrinsic variation, which influences the density fluctuation analysis by imposing a fluctuating signal. This signal is noticed because I have measured non-trivial density fluctuations ($\alpha>0$) at 0 concentration.

### 2.2 Generating background noise

Assume the light is just random noise. At each moment, the noise at one pixel is precisely the same as all other pixels. If pixel intensity is $I_0(t)$ without backgroup noise, then with the noise it would be $I_0(t) + noise(t)$ where $noise(t)$ is a scalar function of time.

Assume the light is fluctuating with time, in a random manner. In each frame, all pixels will add a same noise value. And in the next frame, the noise value will take another random value.

````python
# fluctuation magnitude is 1/10 of image scale (empirical, not value noticeable in experiment)
# use the broadcast feature of np.array, set shape of noise to be (..., 1, 1), where ... is the total number of frames
num_frame = imgstack.shape[0]
background_light = np.random.randint(-10, 10, (num_frame, 1, 1))
imgstack_with_noise = imgstack + background_light
````

````python
gnf = df2(imgstack_with_noise)
plot_gnf(gnf)
plt.show();
````

### 2.3 The role of noise amplitude

In 2.2, I found that amplitudes above 10 lead to very strong fluctuations ($\alpha\approx0.5$). Very low amplitudes, however, can not amplify the fluctuations a lot. In this section, I want to examine the role of noise amplitude. Since I know that amplitude effect plateaus at 10, I will try ``range(1, 10)``.

````python
np.random.seed(1)
num_frame = imgstack.shape[0]
slopeL = []
amplitudes = range(1, 10)
for amplitude in amplitudes:
    background_light = np.random.randint(-amplitude, amplitude, (num_frame, 1, 1))
    imgstack_with_noise = imgstack + background_light
    gnf = df2(imgstack_with_noise)
    ax, slope = plot_gnf(gnf)
    slopeL.append(slope)
plt.show();
````

````python
plt.plot(amplitudes, slopeL)
plt.xlabel('noise amplitude')
plt.ylabel('$\\alpha$')
plt.show();
````

## 3. Smaller scale test

### 3.1 Uniform image

First we assume each frame of the image stack is uniform. They change over time, each pixel changes in the same manner.

````python
L = 2
num_frames = 3
image_low_bound = 0
image_high_bound = 255

imgstack = np.ones((num_frames, L, L))
randnoise = np.random.randint(image_low_bound, image_high_bound, (num_frames, 1, 1))
imgstack *= randnoise
````

````python
num_frames, h, w = imgstack.shape
fig, ax = plt.subplots(1, num_frames, figsize=(8, 2.5))
for i in range(num_frames):
    ax[i].imshow(imgstack[i], cmap='gray', vmin=image_low_bound, vmax=image_high_bound)
    #ax[i].axis('off')
    for j in range(h):
        for k in range(w):          
            ax[i].text(k, j, str(int(imgstack[i, j, k])), color='red', fontdict={'size': 20,  'weight': 'bold' })
    ax[i].set_title('Image ' + str(i))
````

**Standard deviation**

The standard deviation of one pixel over time is identical to all other pixels.

````python
#std1 = randnoise.std()
std1 = np.array([23, 145, 213]).std()
print('one pixel standard deviation: ' + '{0:.2f}'.format(std1))
````

Show the standard deviation field

````python
num_frames, h, w = imgstack.shape
std_field = np.ones((h, w)) * std1
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.imshow(std_field, cmap='gray', vmin=image_low_bound, vmax=image_high_bound)
for j in range(h):
    for k in range(w):          
        ax.text(k, j, '{:.1f}'.format(std_field[j, k]), color='red', fontdict={'size': 15,  'weight': 'bold' })
````

**Varying box size**

Now let's vary the box size from $1$, $2^2$, $3^2$, ..., $L^2$

````python
L = min(imgstack.shape[1:3]) # the smallest dimension of one frame
boxsizes = (np.arange(L) + 1) ** 2 # an array from 1 to L
print(boxsizes)
````

For each box size, we have a cumulative standard deviation corresponding to it.

````python
std_boxwise = std1 * boxsizes # since each pixel has exactly the same std
print(std_boxwise)
````

Plot the cumulative standard deviations as a function of boxsize

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(boxsizes, std_boxwise)
ax.set_xlabel('boxsize [px$^2$]')
ax.set_ylabel('std')
xf, yf, xt, yt, slope = label_slope(boxsizes, std_boxwise, location='n')
plt.plot(xf, yf, ls='--', color='black')
plt.text(xt, yt, '{:.2f}'.format(slope))
plt.xscale('log')
plt.yscale('log')
````

**Why is the slope one?**

Let's use $\Delta N$ to denote the standard deviations and use N to denote the box size.
$$ \Delta N = N \times \Delta N_1 $$
where $\Delta N_1$ is the one pixel temporal standard deviation, which is a constant when each frame is uniform.

### 3.3 Random image

Let's now see why random image results in slope smaller than one.

````python
L = 2
num_frames = 3
image_low_bound = 0
image_high_bound = 255
imgstack = np.random.randint(image_low_bound, image_high_bound, (num_frames, L, L))
````

**Add some noise to the original random image**

Again, we assume noise on each frame is uniform. We will apply the noise as addition and product both.

````python
# Noise as addition
amp = 5
randnoise = np.random.randint(-amp, amp, (num_frames, 1, 1))
imgstack += randnoise
num_frames, h, w = imgstack.shape
fig, ax = plt.subplots(1, num_frames, figsize=(8, 2.5))
for i in range(num_frames):
    ax[i].imshow(imgstack[i], cmap='gray', vmin=image_low_bound, vmax=image_high_bound)
    #ax[i].axis('off')
    for j in range(h):
        for k in range(w):          
            ax[i].text(k, j, str(int(imgstack[i, j, k])), color='red', fontdict={'size': 20,  'weight': 'bold' })
    ax[i].set_title('Image ' + str(i))
````

**Temporal standard deviation**

````python
std_field = imgstack.std(axis=0)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('stdev field')
ax.imshow(std_field, cmap='gray', vmin=image_low_bound, vmax=image_high_bound)
for j in range(h):
    for k in range(w):          
        ax.text(k, j, '{:.1f}'.format(std_field[j, k]), color='red', fontdict={'size': 15,  'weight': 'bold' })
````

````python
L = min(imgstack.shape[1:3]) # the smallest dimension of one frame
boxsizes = (np.arange(L) + 1) ** 2 # an array from 1 to L
print(boxsizes)
````

````python
std_boxwise = np.array([std_field[0, 0], std_field.sum()])
print(std_boxwise)
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(boxsizes, std_boxwise)
ax.set_xlabel('boxsize [px$^2$]')
ax.set_ylabel('std')
xf, yf, xt, yt, slope = label_slope(boxsizes, std_boxwise, location='n')
plt.plot(xf, yf, ls='--', color='black')
plt.text(xt, yt, '{:.2f}'.format(slope))
plt.xscale('log')
plt.yscale('log')
````

**Why is it not one?**

$$ \Delta N \ne N \times \Delta N_1 $$

- $\Delta N_1$ is no longer uniform in space, because of the randomness introduced by the spatial variation in each frame. 

## 4. Systematic investigation of spatial and temporal variation amplitude

### 4.1 Choose good dimensions

````python
num_frames, h, w = 100, 100, 100
amp_i = 100 # image amplitude
amp_n = 1 # noise amplitude
imgstack = np.random.randint(-amp_i, amp_i, (num_frames, h, w))
noise = np.random.randint(-amp_n, amp_n, (num_frames, 1, 1))
imgstack_add_n = imgstack + noise
imgstack_prod_n = imgstack * (noise / amp_i + 1)

gnf = df2(imgstack_prod_n)
plot_gnf(gnf)
````

````python
num_frames, h, w = 100, 100, 100
amp_i = 1 # image amplitude
amp_n = 100 # noise amplitude
imgstack = np.random.randint(-amp_i, amp_i, (num_frames, h, w))
noise = np.random.randint(-amp_n, amp_n, (num_frames, 1, 1))
imgstack_add_n = imgstack + noise
imgstack_prod_n = imgstack * (noise / amp_i + 1)

gnf = df2(imgstack_prod_n)
plot_gnf(gnf)
````

**Comment:** the results above show two combinations of (amp_i, amp_n). When amp_i >> amp_n (1st case), the density fluctuation exponent $\alpha\approx 0$. When amp_i << amp_n (2nd case), the density fluctuation exponent $\alpha\approx 0.5$. These results suggest:
- the dimensions of imgstack (100, 100, 100) is sufficiently large to give good statistics
- the ranging the relative amplitude between image and noise from 0.01 to 100 is large enough to cover both limits
- whether using addition noise or product noise doesn't matter

### 4.2 Amplitude parameter space

Here, let's range amp_i and amp_n from 1 to 100 and plot $\alpha$ in the amplitude parameter space.

````python
num_frames, h, w = 100, 100, 100
# image amplitude
# noise amplitude
amp_iL = range(1, 101)
amp_nL = range(1, 101)
alpha = np.zeros((len(amp_iL), len(amp_nL)))
for i, amp_i in enumerate(amp_iL):
    for j, amp_n in enumerate(amp_nL):
        imgstack = np.random.randint(-amp_i, amp_i, (num_frames, h, w))
        noise = np.random.randint(-amp_n, amp_n, (num_frames, 1, 1))
        imgstack_add_n = imgstack + noise
        gnf = df2(imgstack_add_n)
        x = gnf.n / 100
        y = gnf.d / x ** 0.5
        y = y / y.iat[0]
        xf, yf, xt, yt, slope = label_slope(x, y)
        alpha[i, j] = slope
        print('amp_i = {0}, amp_n = {1}: alpha = {2}'.format(amp_i, amp_n, slope))
````

````python
np.savetxt('temp_data/background_light_noise_alpha.txt', alpha, fmt='%.3f')
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.imshow(alpha)
ax.set_xlabel('noise amplitude')
ax.set_ylabel('light amplitude')
mappable = mpl.cm.ScalarMappable(norm=mpl.colors.Normalize(vmin=0, vmax=0.5), cmap='viridis')
plt.colorbar(mappable=mappable, ax=ax)
````
